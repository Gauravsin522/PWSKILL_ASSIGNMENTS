```python
#Question5: What are the key properties of the Poisson distribution, and when is it appropriate to use this distribution?  

#Answer:- The Poisson distribution is a discrete probability distribution that models the number of events occurring in a fixed
#interval of time or space if these events occur with a known constant mean rate and independently of the
#time since the last event. The key properties of the Poisson distribution are
#1. The Poisson distribution is a discrete distribution, meaning it can only take on non-negative
#integer values.
#2. The Poisson distribution is a one-parameter distribution, meaning it is characterized by a single
#parameter, λ (lambda), which is the average rate of events occurring in the fixed interval.
#3. The Poisson distribution is a memoryless distribution, meaning the probability of an event occurring
#does not depend on the time since the last event.
#4. The Poisson distribution is a skewed distribution, meaning it is not symmetric about its mean
#The Poisson distribution is appropriate to use when the following conditions are met:
#1. The number of events occurring in the fixed interval is small compared to the size of the
#population.
#2. The events occur independently of each other.
#3. The events occur at a constant rate over the fixed interval.
#4. The events are rare, meaning the probability of an event occurring is small.
#5. The events are independent and identically distributed.
#6. The fixed interval is well-defined and finite
#7. The average rate of events (λ) is known and constant.
#8. The number of events is countable and non-negative.
#9. The events are not influenced by external factors such as time, location, or other events
#10. The Poisson distribution is a good approximation to the binomial distribution when the number of
#trials (n) is large and the probability of success (p) is small.
```

```python
#Question6: Define the terms "probability distribution" and "probability density function" (PDF). How does a PDF differ from a probability mass function (PMF)?   

# Answer :- 
# A probability distribution is a function that describes the probability of each possible outcome in a sample space.
# A probability density function (PDF) is a function that describes the probability distribution of a continuous random
# variable. It is a function that satisfies the following properties: (1) f(x) ≥
# 0 for all x in the domain, (2) ∫f(x)dx =
# 1, and (3) ∫f(x)dx = P(a ≤ X ≤
# b) for any interval [a, b].
# A probability mass function (PMF) is a function that describes the probability distribution of a discrete
# random variable. It is a function that satisfies the following properties: (1) p(x)
# 0 for all x in the domain, (2) ∑p(x) =
# 1, and (3) p(x) = P(X =
# x) for any x in the domain.
# A PDF differs from a PMF in that a PDF is used for continuous random variables, while
# a PMF is used for discrete random variables.
# In a PDF, the probability of an outcome is given by the area under the curve, while
# in a PMF, the probability of an outcome is given by the value of the function at
# that outcome  

# Example: Consider a continuous random variable X with a PDF f(x) = 1/2
# for 0 ≤ x ≤ 1 and f(x) = 0 otherwise. The probability
# of X being between 0.5 and 1 is given by the area under the curve between these two points.
# The probability of X being exactly 0.5 is given by the value of the function at
# that point, which is 1/2.
# In contrast, consider a discrete random variable Y with a PMF p(y) = 1
# for y = 1 and p(y) = 0 otherwise. The probability of Y being
# 1 is given by the value of the function at that point, which is 1.
# The probability of Y being between 0.5 and 1 is given by the sum of
# the probabilities of the outcomes in that interval, which is 0.
# In summary, a PDF is used for continuous random variables and describes the probability distribution
# of a continuous random variable, while a PMF is used for discrete random variables and describes the
# probability distribution of a discrete random variable.
# In Python, we can use the scipy.stats module to create a probability distribution and calculate the probability
# of an outcome. For example
from scipy.stats import norm
import numpy as np
# Create a normal distribution with mean 0 and standard deviation 1
dist = norm(loc=0, scale=1)
# Calculate the probability of X being between 0 and 1
prob = dist.cdf(1) - dist.cdf(0)
# Output: 0.6826894921370859
# Calculate the probability of X being exactly 0.5
prob = dist.pdf(0.5)
# Output: 0.3989422804014327
```

```python
# Question7: Explain the Central Limit Theorem (CLT) with example.   

#Answer :- 
# The Central Limit Theorem (CLT) states that the distribution of the mean of a large
# sample of independent and identically distributed (i.i.d.) random variables will be
# approximately normally distributed, regardless of the original distribution of the
# individual variables.
# This theorem is a fundamental concept in statistics and is widely used in many fields,
# including finance, engineering, and social sciences.
# Here is an example of how the CLT works:
# Suppose we have a population of exam scores with a mean of 70 and a standard deviation of
# 10. We take a random sample of 100 exam scores from this population and calculate the
# mean of the sample. We repeat this process many times and plot the distribution of the
# sample means. We would expect the distribution of the sample means to be approximately
# normally distributed, with a mean of 70 and a standard deviation of 10/sqrt(
# 100) = 1.
import numpy as np
import matplotlib.pyplot as plt
# Define the population parameters
mu = 70
sigma = 10
# Define the sample size
n = 100
# Generate a large number of random samples from the population
num_samples = 10000 
# Initialize an empty list to store the sample means
sample_means = []
for i in range(num_samples):
    # Generate a random sample from the population
    sample = np.random.normal(mu, sigma, n)
    
    # Calculate the mean of the sample
    sample_mean = np.mean(sample)
    # Append the sample mean to the list
    sample_means.append(sample_mean)
    # Plot the histogram of the sample means
    plt.hist(sample_means,bins=30, density=True)
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')
    plt.show()
    # The histogram of the sample means will be approximately normally distributed, with a mean of 70 and   
    # a standard deviation of 1. This demonstrates the Central Limit Theorem.
             
```

```python
#Question9: Given a sample mean of 105, a population mean of 100, a standard deviation of 15, and a sample 
#size of 25, calculate the z-score and p-value. Based on a significance level of 0.05, do you reject or fail to 
#reject the null hypothesis? 
#Task: Write Python code to calculate the z-score and p-value for the given data. 
#Objective: Apply the formula for the z-score and interpret the p-value for hypothesis testing.   

import scipy.stats as stats 
import numpy as np
sample_mean = 105
population_mean = 100
std_dev = 15
sample_size = 25

# Calculate the z-score
z_score = (sample_mean - population_mean) / (std_dev / (sample_size ** 0.5))

# Calculate the p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")

significance_level = 0.05
if p_value < significance_level:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
#Question10: Simulate a binomial distribution with 10 trials and a probability of success of 0.6 using Python. 
#Generate 1,000 samples and plot the distribution. What is the expected mean and variance? 
#Task: Use Python to generate the data, plot the distribution, and calculate the mean and variance. 
#Objective: Understand the properties of a binomial distribution and verify them through simulation. 
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n_trials = 10
p_success = 0.6
n_samples = 1000

# Generate 1,000 samples from the binomial distribution
samples = np.random.binomial(n_trials, p_success, n_samples)

# Plot the distribution
plt.hist(samples, bins=np.arange(n_trials + 2) - 0.5, density=True, edgecolor='black')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution (n=10, p=0.6)')
plt.show()

# Calculate the expected mean and variance
expected_mean = n_trials * p_success
expected_variance = n_trials * p_success * (1 - p_success)

# Calculate the sample mean and variance
sample_mean = np.mean(samples)
sample_variance = np.var(samples)

# Output the results
print(f"Expected Mean: {expected_mean}")
print(f"Expected Variance: {expected_variance}")
print(f"Sample Mean: {sample_mean}")
print(f"Sample Variance: {sample_variance}")
```
