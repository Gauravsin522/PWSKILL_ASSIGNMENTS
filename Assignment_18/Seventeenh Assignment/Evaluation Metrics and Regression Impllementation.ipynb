# Theoretical

```python
# 1. What does R-squared represent in a regression model?

# Answer: R-squared (RÂ²) indeed represents the proportion of the variance in the dependent variable that is predictable from the independent variable(s).
# It's a measure of how well the regression model fits the data
```

```python
# 2. What are the assumptions of linear regression ?

# Answer: The assumptions of linear regression are:
# 1. Linearity: The relationship between the independent and dependent variables is linear.
# 2. Independence: The residuals are independent.
```

```python
# 3. What is the difference between R-squared and Adjusted R-squared?

# Answer: R-squared is a measure of how well the regression model fits the data, while Adjusted R-squared is a modified version
# of R-squared that has been adjusted for the number of predictors in the model.
```

```python
# 4. Why do we use Mean Squared Error (MSE)?

# Answer: Mean Squared Error (MSE) is a measure of the average squared difference between the actual and predicted values in a regression model.
# It is used to evaluate the performance of the model and to compare different models.
```

```python
# 5. What does an Adjusted R-squared value of 0.85 indicate?

# Answer: An Adjusted R-squared value of 0.85 indicates that 85% of the variance in the dependent variable is predictable from the independent variable(s)
# in the regression model, after adjusting for the number of predictors in the model. It suggests that the model fits the data well.
```

```python
# 6. How do we check for normality of residuals in linear regression?

# Answer: We can check for normality of residuals in linear regression by plotting a histogram of the residuals and checking if it follows a normal distribution.
# We can also use statistical tests like the Shapiro-Wilk test or the Kolmogorov-Smirnov test to test for normality.
```

```python
# 7. What is multicollinearity, and how does it impact regression?

# Answer: Multicollinearity is a phenomenon in which two or more independent variables in a regression model are highly correlated with each other.
# It can impact regression by making it difficult to estimate the individual effects of the independent variables on the dependent variable,
# leading to unstable and unreliable estimates of the coefficients.
```

```python
# 8. What is Mean Absolute Error (MAE)?

# Answer: Mean Absolute Error (MAE) is a measure of the average absolute difference between the actual and predicted values in a regression model.
# It is used to evaluate the performance of the model and to compare different models.
```

```python
# 9. What are the benefits of using an ML pipeline?

# Answer: The benefits of using an ML pipeline include:
# 1. Automation: ML pipelines automate the process of training, testing, and deploying machine learning models, saving time and effort.
# 2. Reproducibility: ML pipelines make it easy to reproduce the results of a machine learning experiment by capturing the entire workflow.
# 3. Scalability: ML pipelines can be easily scaled to handle large datasets and complex machine learning models.
# 4. Flexibility: ML pipelines allow for easy experimentation with different algorithms, hyperparameters, and data preprocessing steps.
```

```python
# 10. Why is RMSE considered more interpretable than MSE?

# Answer: RMSE (Root Mean Squared Error) is considered more interpretable than MSE (Mean Squared Error) because it is in the same units as the dependent variable,
# making it easier to interpret the error in the context of the problem. RMSE is the square root of the average squared difference between the actual and predicted values,
# while MSE is the average squared difference between the actual and predicted values.
```

```python
# 11. What is pickling in Python, and how is it useful in ML?

# Answer: Pickling in Python is the process of serializing and deserializing objects to and from a byte stream. It is useful in ML for saving trained machine learning models
# to disk so that they can be reused later without having to retrain the model. Pickling allows for easy storage and retrieval of machine learning models.
```

```python
# 12. What does a high R-squared value mean?

# Answer: A high R-squared value indicates that a large proportion of the variance in the dependent variable is predictable from the independent variable(s)
# in the regression model. It suggests that the model fits the data well and that the independent variables are good predictors of the dependent variable.
```

```python
# 13. What happens if linear regression assumptions are violated?

# Answer: If linear regression assumptions are violated, the estimates of the coefficients may be biased and inefficient, and the predictions made by the model may be unreliable.
# Violations of the assumptions can lead to incorrect inferences and poor model performance.
```

```python
# 14. How can we address multicollinearity in regression?

# Answer: Multicollinearity in regression can be addressed by:
# 1. Dropping one of the highly correlated variables.
# 2. Using regularization techniques like Lasso or Ridge regression.
# 3. Combining the correlated variables into a single composite variable.
# 4. Collecting more data to reduce the impact of multicollinearity.
```

```python
# 15. How can feature selection improve model performance in regression analysis?

# Answer: Feature selection can improve model performance in regression analysis by reducing the complexity of the model and removing irrelevant or redundant features.
# This can help prevent overfitting, improve the interpretability of the model, and reduce the computational cost of training the model.
```

```python
# 16. How is Adjusted R-squared calculated?

# Answer: Adjusted R-squared is calculated using the formula:
# Adjusted R-squared = 1 - (1 - R-squared) * (n - 1) / (n - k - 1)
```

```python
# 17. Why is MSE sensitive to outliers?

# Answer: MSE (Mean Squared Error) is sensitive to outliers because it squares the errors between the actual and predicted values, giving more weight to large errors.
# Outliers can have a disproportionate impact on the MSE, leading to biased estimates of the model parameters and poor model performance.
```

```python
# 18. What is the role of homoscedasticity in linear regression?

# Answer: Homoscedasticity is the assumption that the variance of the residuals is constant across all levels of the independent variables in a linear regression model.
# It is important for the validity of the regression model and the interpretation of the coefficients. Violations of homoscedasticity can lead to biased and inefficient estimates.
```

```python
# 19. What is Root Mean Squared Error (RMSE)?

# Answer: Root Mean Squared Error (RMSE) is a measure of the average squared difference between the actual and predicted values in a regression model.
# It is the square root of the average squared difference between the actual and predicted values, and it is used to evaluate the performance of the model and to compare different models.
```

```python
# 20. Why is pickling considered risky?

# Answer: Pickling is considered risky because it can be a security risk if the pickled objects are unpickled from untrusted sources.
# Unpickling maliciously crafted pickled objects can lead to code execution vulnerabilities, data leakage, and other security issues.
# It is important to only unpickle objects from trusted sources and to validate the pickled objects before unpickling them.
```

```python
# 21. What alternatives exist to pickling for saving ML models?

# Answer: Alternatives to pickling for saving ML models include:
# 1. Serialization formats like JSON or XML.
# 2. Saving the model weights and architecture separately.
# 3. Using specialized model serialization libraries like joblib or dill.
# 4. Saving the model as a PMML (Predictive Model Markup Language) file.
```

```python
# 22. What is heteroscedasticity, and why is it a problem?

# Answer: Heteroscedasticity is the phenomenon in which the variance of the residuals is not constant across all levels of the independent variables in a linear regression model.
# It is a problem because it violates the assumption of homoscedasticity, leading to biased and inefficient estimates of the coefficients, incorrect inferences, and unreliable predictions.
```

```python
# 23. How can interaction terms enhance a regression model's predictive power?

# Answer: Interaction terms can enhance a regression model's predictive power by capturing the combined effect of two or more independent variables on the dependent variable.
# They allow the model to account for non-additive relationships between the variables and can improve the model's ability to predict the dependent variable accurately.
```

```python
# 1. Write a Python script to visualize the distribution of errors (residuals) for a multiple linear regression model using Seaborn's "diamonds" dataset. 

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

df=sns.load_dataset('diamonds')

print(df.head())

X=df[['carat','depth','table','x','y','z']]
y=df['price']

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=1,test_size=0.2)

model=LinearRegression()

# fit the model
model.fit(X_train,y_train)

# predict the values
y_pred=model.predict(X_test)

# calculate the residuals
residuals=y_test-y_pred

# plot the distribution of residuals
plt.figure(figsize=(10,6))
sns.histplot(residuals, kde=True , bins=50)
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Distribution of Residuals')
plt.show()

```

```python
# 4. Write a Python script that creates a machine learning pipeline with feature scaling and evaluates the performance of different regression models 

from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression,Ridge,ElasticNet,Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split,cross_val_score

df = sns.load_dataset("tips")

X = df["total_bill"].values.reshape(-1, 1)
y = df["tip"].values

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, test_size=0.2)

model={
    "Linear":LinearRegression(),
    "Ridge":Ridge(alpha=1.2),
    "Lasso":Lasso(1.5),
    "Elastic":ElasticNet(l1_ratio=0.5)
}

for name,model in model.items():
    pipeline=Pipeline([
        ("Scaler",StandardScaler()),
        ("Regression",model)
                      
    ])
    
    pipeline.fit(X_train,y_train)
    y_pred=pipeline.predict(X_test)
    
    r2=r2_score(y_test,y_pred)
    
    cv=cross_val_score(pipeline,X,y,cv=5)
    
    print(f"{name}:")
    print(f"R-squared score: {r2}")
    print(f"Cross-validation scores: {cv}")
    print(f"Mean CV score: {cv.mean()}")
    print("-" * 30)
```

```python
# 11. Write a Python script that fits polynomial regression models of different degrees to a synthetic dataset and 
# compares their performance. 

n_samples=1000

X = 2 * np.random.rand(n_samples, 1)
y = 4 + 3 * X + np.random.randn(n_samples, 1)

df=pd.DataFrame({"feature":X.flatten(),"target":y.flatten()})

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, test_size=0.2)

from sklearn.preprocessing import PolynomialFeatures

def fit_and_evaluate_polynomial(degree):
    poly=PolynomialFeatures(degree=degree)
    X_train_poly=poly.fit_transform(X_train)
    X_test_poly=poly.transform(X_test)
    
    model=LinearRegression()
    model.fit(X_train_poly,y_train)
    
    y_pred=model.predict(X_test_poly)
    
    mse=mean_squared_error(y_test,y_pred)
    r2=r2_score(y_test,y_pred)
    return mse,r2

degrees=[1,2,3,4,5]
result={}

for degree in degrees:
    mse,r2=fit_and_evaluate_polynomial(degree)
    result[degree]={"MSE":mse,"R_squared":r2}
    
for degree, metrics in result.items():
    print(f"Degree: {degree}")
    print(f"Mean Squared Error: {metrics['MSE']}")
    print(f"R-squared: {metrics['R_squared']}")
    print()
```

```python
# 21. Write a Python script that demonstrates how to serialize and deserialize machine learning models using 
# joblib instead of pickling. 

np.random.seed(1)
X = np.random.rand(100, 5)
y = np.random.rand(100)

X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=1,test_size=0.2)

model=LinearRegression()
model.fit(X_train,y_train)

y_pred=model.predict(X_test)

import joblib

# serialize 

joblib_file="Linear_Regression_model.joblib"
joblib_dump=joblib.dump(model,joblib_file)

print(f"Model serialized to {joblib_file}")

# deserialize

loaded_data=joblib.load(joblib_file)

y_pred=loaded_data.predict(X)

print(y_pred)
```

```python
# 25. Write a Python script that compares polynomial regression models of different degrees and prints the R-
# squared score for each.

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

def compare_degree(degree):
    poly = PolynomialFeatures(degree=degree)
    X_train_poly = poly.fit_transform(X_train)
    X_test_poly = poly.transform(X_test)
    
    model = LinearRegression()
    model.fit(X_train_poly, y_train)
    
    y_pred = model.predict(X_test_poly)
    
    r2 = r2_score(y_test, y_pred)
    
    return r2

degrees = [1, 2, 3, 4, 5]
result = {}

for degree in degrees:
    r2 = compare_degree(degree)
    result[degree] = {"R_squared": r2}

for degree, metrics in result.items():
    print(f"Degree: {degree}")
    print(f"R-squared: {metrics['R_squared']}")
```
